```python
# number of bytes =size * itemsize
import numpy as np 
a=np.array([1,2,3],dtype='int16')
b=np.array([3,4,5])
print("size of an array a in bytes: ",a.nbytes)
print("size of an array b in bytes: ",b.nbytes) # number of bytes =size * itemsize
print("a", a.itemsize*a.size)
print("b",b.itemsize*b.size)
```

```python
# functions that can be performed on matrix

import numpy as np
max=np.random.randint(1,10,size =(4,4)) 
print(max)
det=np.linalg.det(max)
print("determinant: ",det)
trace=np.linalg.trace(max)
print("trace: ",trace)
transpose=np.linalg.matrix_transpose(max)
print("transpose: ",transpose)
rank=np.linalg.matrix_rank(max)
print("rank: ",rank)
diagonal=np.linalg.diagonal(max)
print("diagonal: ",diagonal)
singular=np.linalg.vector_norm(max)
print(singular)
```
