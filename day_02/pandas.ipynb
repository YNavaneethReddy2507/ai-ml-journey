ACCESSING DATA WITH PANDAS

```python
coffee.head(10)
```

```python
coffee.tail(5)
```

```python
coffee.sample(10) # randomly select the data among dataset in the csv file 
```

```python
coffee.loc[[1,7]] # we can select the data by using loc and passing the index of the data we want to select
```

```python
coffee.loc[5:10] # we can use slicing to retrieve the data from index ranges
```

```python
coffee.loc[5:9,["Day","Units Sold"]] # we can use slicing to retrieve the data from index ranges and also we can pass the column name to get the specific column data
```

```python
coffee.iloc[:,[0,2]] # [[selective rows index],[selective column index]]
```

```python
# we can modify the data by using loc and iloc and passing the index of the data we want to modify and also we can pass the column name to get the specific column data and then we can assign the new value to it
coffee.loc[1,"Units Sold"]=20
coffee.head()
```

```python
# at-> gives an single value based on the reqiured index and column name
coffee.at[1,"Units Sold"] 
```

```python
# iat-> gives an single value based on the reqiured index and column index
coffee.iat[0,1]
```

```python
coffee.sort_values(["Units Sold","Coffee Type"], ascending=True) # we can sort the values by using sort_values and passing the column name and also we can pass the ascending as False to sort in descending order
```

```python
for index,row in coffee.iterrows():
    print(row['Units Sold'])
    print(index)
    print("\n")
```
